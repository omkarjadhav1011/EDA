## One Hot Encoding

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_excel('Data/data.xlsx')
df
```

```python
df['Country'].unique()
```

```python
ohe = OneHotEncoder()
```

```python
ohe
```

```python
fa = ohe.fit_transform(df[['Colour','Country']]).toarray()
```

```python
fa
```

```python
fa.size,fa.shape
```

```python
f_label = ohe.categories_
f_label
```

```python
f_label2 = np.array(f_label).ravel()
f_label2
```

```python
df1 = pd.DataFrame(fa,columns=f_label2)
df1
```
