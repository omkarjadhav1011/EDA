### <span style="color: green"> Column Transformation </span>

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("ESD.xlsx")   #ESD data set
df
```

```python
df.columns
```

```python
df.loc[(df["Bonus %"]) > 0, "Get Bonus"] = "Bonus"
df.loc[(df["Bonus %"]) == 0, "Get Bonus"] = "No Bonus"
df
```

```python
df.columns
```

```python
df["Get Bonus"].value_counts()
```

```python
df1 = pd.read_excel("practice.xlsx")      #practice data set
display(df1)
```

```python
df1.columns
```

```python
df1["Full Name"] = df1['First name']+' '+df1['Last name']
df1
```

```python
df1["Full Name"] = df1['First name'].str.capitalize()+' '+df1['Last name'].str.capitalize()
df1
```

```python
df1["Bonus"] = df1["Salary"]*0.10
df1
```

```python
df1.rename(columns={'EEID':"Emp-ID", "First name": "First Name", "Last name": "Last Name"}, inplace = True)
df1
```
