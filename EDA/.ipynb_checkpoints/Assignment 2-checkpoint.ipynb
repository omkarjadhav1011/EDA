#### 1. Column Transformation

```python
import pandas as pd
import numpy as np
```

```python
# importing ESD.xlsx dataset
df_esd = pd.read_excel("Data/ESD.xlsx")
df_esd.head()
```

```python
# This code is used to create new column in dataset based on previous one 
df_esd.loc[df_esd["Bonus %"]>0, 'Get Bouns'] = "Yes"
df_esd.loc[df_esd['Bonus %']==0, 'Get Bouns']= "No"
df_esd.head()
```

```python
# Importing practice dataset
df_pra = pd.read_excel("Data/practice .xlsx")
df_pra.head()
```

```python
# Create a new column 'Full Name'(Dataset:-practice.xlsx)
df_pra["Full Name"] = df_pra['First name'].str.capitalize()+" "+df_pra['Last name'].str.capitalize()
df_pra.head()
```

```python
# Create a new column 'Bonus'(Dataset:-practice.xlsx)
df_pra['Bouns'] = df_pra['Salary']*0.10
df_pra.head()
```

```python
# Rename Column EEID to EmpID (Dataset:-practice.xlsx)
df_pra.rename(columns={'EEID':'EmpID'})
```

```python
# Rename Columns Hire Date to Hire-Date, Exit Date to Exit-Date (Dataset:-ESD.xlsx)
df_esd.rename(columns={'Hire Date' : "Hire-Date",'Exit Date':'Exit-Date'}).head()
```

#### 3. Use groupby function in dataset (Dataset:-ESD.xlsx)

```python
# Department wise employee count using ‘EEID’ column.
df_esd.groupby("Department").aggregate({'EEID':'count'})
```

```python
# Department, Gender wise employee count using EEID column.
df_esd.groupby(["Department","Gender"]).aggregate({'EEID':'count'})
```

```python
df_esd.groupby("Country").aggregate({'Annual Salary':"min"})
```

```python
df_esd.groupby(["Country","Gender"]).aggregate({'Annual Salary':"max"})
```

```python
df_esd.groupby(["Country","Gender","Department"]).aggregate({'Annual Salary':"max"})
```

```python
df_esd.groupby(["Country","Gender","Department"]).aggregate({'Annual Salary':"max",'Age':"min"})
```

```python
df_esd.groupby(["Country","Gender"]).aggregate({'Annual Salary':"min","Age":"min"})
```

#### 4. Merge Data Frames

```python
data_1 = {"Emp_ID": ['E01', 'E07', 'E03', 'E08', 'E05', 'E06'],
         "Name": ["Abc", "Def", "Ghi", "Jkl", "Mno", "Jkl"],
         "Age": [23, 24, 45, 35, 27, 30]}
data_2 = {"Emp_ID": ['E01', 'E02', 'E03', 'E04', 'E05', 'E10'],
         "Salary": [25000, 30000, 60000, 55000, 70000, 85000]}
```

```python
df_for_merge1 = pd.DataFrame(data_1)
df_for_merge1
```

```python
df_for_merge2 = pd.DataFrame(data_2)
df_for_merge2
```

```python
df_merge = pd.merge(df_for_merge1,df_for_merge2,
                   how='left',
                   on='Emp_ID')
df_merge
```
