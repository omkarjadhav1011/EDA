# Handling duplicate values

```python
data = {'Name' : ['Ram','Sham',"sita",'Abhi','Om','Sangram','Sai','Ram','Abhi'],
       'Mark':[10,9,7,1,0,10,10,10,4],
       'Course':['ds','aiml','cse','entc','ds','civil','mech','ds','cse']}
```

```python
df7=pd.DataFrame(data)
df7
```

```python
df7.duplicated()
```

```python
df7.duplicated().sum()
```

```python
df7.loc[df7.duplicated()]
```

```python
df8 = df7.loc[df7.Name.duplicated()]
df8
```

```python
df9 = df7.loc[df7.Course.duplicated()]
df9
```

```python
df9 = df7.loc[df7.Mark.duplicated()]
df9
```

```python
df10 = df7.drop_duplicates(subset='Name',keep='first') 
df10
```

```python
df11 = df7.drop_duplicates(subset="Name",keep='last')
df11
```

```python
df12  = df7.drop_duplicates(subset='Name',keep=False)
```
